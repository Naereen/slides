```python
import numpy as np

import matplotlib.pyplot as plt
import matplotlib.ticker as mtick

import seaborn as sns
sns.set(context="notebook", style="white", palette="hls", font="sans-serif", font_scale=1.3)
```

```python
base_percent_S = np.array([0.3,0.2,0.1,0.1,0.05,0.05,0.02,0.08,0.01,0.09])
S1K = lambda S: S * base_percent_S
```

```python
SplusD = 10_000
```

```python
S1K_fromD = lambda percentD: S1K((1-percentD)*SplusD)
```

```python
S1K_fromD(0.10)
```

```python
plt.xticks
```

```python
def hist_of_S1K(valuesS, title=None):
    K = len(valuesS)
    valuesK = np.arange(1, K+1)
    plt.figure(dpi=130, figsize=(10, 7))
    plt.bar(valuesK, valuesS / np.sum(valuesS), color="darkgreen")
    plt.xticks(valuesK)
    plt.xlabel(r"Radio channel $k=1,\dots,{}$".format(K))
    plt.ylabel("Percentage of static devices in this channel")
    if title: plt.title(title)
    ax = plt.gca()
    ax.yaxis.set_major_formatter(mtick.PercentFormatter(1.0))
    plt.show()
```

```python
percentD = 0.10
K = len(base_percent_S)
nb_S = int((1-percentD)*SplusD)

hist_of_S1K(S1K_fromD(percentD), "Repartition of at most $S={}$ static devices".format(SplusD))

with plt.xkcd():
    percentD = 0.10
    K = len(base_percent_S)
    nb_S = int((1-percentD)*SplusD)

    hist_of_S1K(S1K_fromD(percentD), "Repartition of at most $S={}$ static devices".format(SplusD))
```
